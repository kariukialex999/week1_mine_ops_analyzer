1. 1. KPI FUNCTION: Write a Function calculate_efficiency(actual_output, target_output) that returns the efficiency percentage.

```python
def calculate_efficiency(actual_output, target_output):
    return (actual_output / target_output) * 100
```

```python
calculate_efficiency(80, 100)
```

2. Status Classifier: Write a function get_operational_status(effieciency) that returns:
    "Critical" if efficiency < 70%
    "Warning" if efficiency is between 70% and 89%
    "Noraml" if efficiency >= 90%.

```python
def get_operational_status(efficiency):
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:
        return "Warning"
    else:
        return "Normal" 
```

```python
get_operational_status(99)
```

3. Data Modeling: Create a dictionary representing 3 different operational sites. Each site should should have keys for name, actual_output and target_output.

```python
#Data model representing 3 operational sites in the logistics industry

operational_sites = {
    "Nairobi Distribution Hub": {
        "actual_output": 14200,
        "target_output": 15000,
    },
    "Mombasa Port Depot": {
        "actual_output": 6500,
        "target_output": 9500,
    },
    "Kisumu Fulfillment Centre": {
        "actual_output": 5100,
        "target_output": 6000,
    },
}
```

4. Execution: Use a loop to iterate through your 3 sites, calculate their efficiency, determine their status, and print a formatted report for each.

```python
print("=" * 40)
print("Operational Efficiency Report")
print("=" * 40)

for code, sites in operational_sites.items(): 
    eff = calculate_efficiency(sites["actual_output"], sites["target_output"])
    status = get_operational_status(eff)
    print(f"\nSITE       : {code}")
    print(f"Output     : {sites['actual_output']:,} / {sites['target_output']:,}")
    print(f"Efficiency : {eff:.1f}%")
    print(f"Status     : {status}")
    print("-" * 40)
```
